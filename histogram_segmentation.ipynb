# Practica calculo de histogramas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

im_gray = cv2.imread('./images/cameraman_remastered.png', cv2.IMREAD_GRAYSCALE)


# Visualizacin de la imagen
plt.figure(figsize=(8,8))
plt.imshow(im_gray, cmap='gray')
plt.title('Imagen de Cameraman')
plt.axis('off')
plt.show()
```

```python

# Calculo de histograma
plt.figure(figsize=(6,4))
plt.hist(im_gray.ravel(), 256, [0,256])
plt.title('Histograma de la Imagen de Monedas')
plt.xlabel('Intensidad de los pxeles')
plt.ylabel('Frecuencia')
plt.show()
```
