```python
# Basic operations 

#  Desarrolla una función capaz de recortar y pegar un fragmento rectangular de una imagen en otra ubicación (de la misma imagen).
# Para esto, la función debería recibir como parámetros de entrada: 
# la imagen (I), las coordenadas de la esquina superior izquierda del rectángulo (x_origen, y_origen), 
# su ancho (w) y alto (h). Finalmente, la función también recibe las coordenadas de destino (x_destino, y_destino), 
# que indican donde se pegará el fragmento (esquina superior izquierda).
```

```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt

im_gray = cv2.imread('./images/cameraman.tif', cv2.IMREAD_UNCHANGED)


print(f'Forma de la imagen: {im_gray.shape}')
print(f'Tipo de datos: {im_gray.dtype}')
```

```python
plt.imshow(im_gray, cmap='gray')
plt.title('Imagen Original')
#plt.axis('off')
plt.show()
```

```python



def copy_section(image_original, origin, dimentions, target, origin_values = 255 ):

    image = image_original.copy()
    [x_origin, y_origin] = origin
    [w,h] = dimentions
    [x_target, y_target ] = target

    chunk_values = image[y_origin:y_origin+h,x_origin:x_origin +w]

    image[y_target:y_target+h,x_target:x_target +w] = chunk_values
    image[y_origin:y_origin+h,x_origin:x_origin +w] = origin_values

    return image

    
    
```

```python

origin = [80,40]
chunk = [50,50]
target = [150,150]

img = copy_section(im_gray, origin, chunk, target, 255)

plt.imshow(img, cmap='gray')
plt.title('Imagen alterada')
#plt.axis('off')
plt.show()
```
