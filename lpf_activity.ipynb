Seleccione una imagen cualquiera en escala de grises y luego desarrolle un script en Python que permita leer dicha imagen y aplicarle una de las técnicas de filtrado estudiadas en clase (LPF). El script deberá calcular y mostrar el histograma de la imagen tanto antes como después de aplicar el filtrado.



```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

# img = cv2.imread('./images/0009.tif', cv2.IMREAD_GRAYSCALE)
# img = cv2.imread('./images/leaf_1.JPG', cv2.IMREAD_GRAYSCALE)
# img = cv2.imread('./images/test_pattern_blurring_orig.tif', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('./images/coins.png', cv2.IMREAD_GRAYSCALE)


# Get histogram
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Apply a low pass filter (Gaussian blur) (low effect)
kernel_size_l = 21 
sigma_l = 5
w_2_b = cv2.getGaussianKernel(kernel_size_l, sigma_l) * cv2.getGaussianKernel(kernel_size_l, sigma_l).T 
filtered =  cv2.filter2D(img, -1, w_2_b, borderType=cv2.BORDER_REPLICATE)


# Apply a high pass filter (Gaussian blur) (high effect)
kernel_size = 91
sigma = 20
k_high = cv2.getGaussianKernel(kernel_size, sigma) * cv2.getGaussianKernel(kernel_size, sigma).T
filtered_high = cv2.filter2D(img, -1, k_high, borderType=cv2.BORDER_REPLICATE)


# Get histogram of filtered image
hist_filtered = cv2.calcHist([filtered], [0], None, [256], [0, 256])

# Get histogram of filtered image (high pass)
hist_filtered_high = cv2.calcHist([filtered_high], [0], None, [256], [0, 256])


plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.plot(hist, color='black')
plt.xlim([0, 256])
plt.title('Original image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(2, 3, 2)
plt.plot(hist_filtered, color='black')
plt.xlim([0, 256])
plt.title('Blurred Image (soft)')
plt.text(120, hist_filtered.max() - 150, 'sigma = %d' % sigma_l, fontsize=10, color='black')
plt.text(120, hist_filtered.max() - 700, 'kernel size = %d' % kernel_size_l, fontsize=10, color='black')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(2, 3, 3)
plt.plot(hist_filtered_high, color='black')
plt.xlim([0, 256])
plt.title('Blurred Image (hard)')
plt.text(120, hist_filtered_high.max() - 50, 'sigma = %d' % sigma, fontsize=10, color='black')
plt.text(120, hist_filtered_high.max() - 250, 'kernel size = %d' % kernel_size, fontsize=10, color='black')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')


# Show images
plt.subplot(2, 3, 4)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow(filtered, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(filtered_high, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()



# Bode plot of the filter
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(w_2_b, cmap='gray')
plt.title('Low Pass Filter (soft)')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(k_high, cmap='gray')
plt.title('Low pass Filter (hard)')
plt.axis('off')
plt.tight_layout()
plt.show()


```
